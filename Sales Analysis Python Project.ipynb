We can see from the above output that the column 'Quantity Ordered' also have a string value having 355 entries, now we have to drop those value as they are a type of error 

```python
#First creating a boelian mask for entries having Quantity Ordered with will show True for entries having Ordered Quantity 
final_df['Quantity Ordered'].str.contains('Quantity Ordered')
```

```python
#using loc to find those entries

final_df.loc[final_df['Quantity Ordered'].str.contains('Quantity Ordered'),:]
```

So above are the 355 entries which has to filtered from the actual dataframe 

```python
#creating a function vice_versa which interchange the boelian values so that we get entries as True for False and vice_versa

def vice_versa(x):
    if x==True:
        return False
    else:
        return True
```

```python
#Now creating the boolean series for 

final_df['Quantity Ordered'].str.contains('Quantity Ordered').apply(vice_versa)
```

```python
#Now using the above boolean mask we will eliminated entries containing string values in column 'Quantity Ordered'

final_df.loc[final_df['Quantity Ordered'].str.contains('Quantity Ordered').apply(vice_versa),:]
```

```python
final_df = final_df.loc[final_df['Quantity Ordered'].str.contains('Quantity Ordered').apply(vice_versa),:]
```
